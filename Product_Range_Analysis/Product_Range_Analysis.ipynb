```python
#function with looking for similarity
decor_nlp = []
accessories_nlp = []
kitchenware_nlp = []
furniture_nlp = []
stationary_nlp = []
christmas_nlp = []
toys_games_nlp = []

for item in decor:
    decor_nlp.append(nlp(item))

for item in accessories:
    accessories_nlp.append(nlp(item))

for item in kitchenware:
    kitchenware_nlp.append(nlp(item))

for item in furniture:
    furniture_nlp.append(nlp(item))

for item in stationary:
    stationary_nlp.append(nlp(item))

for item in toys_games:
    toys_games_nlp.append(nlp(item))

for item in christmas:
    christmas_nlp.append(nlp(item))
```

```python
# descriptions=goods['description'].unique()
# descriptions=pd.DataFrame(descriptions, columns=['description'])
# descriptions
unique_desc_df=pd.DataFrame(goods['description'].unique(), columns=['description'])
unique_desc_df.sample(5)
```

```python
#items - manually generated list
def generate_auto_list(items):
  auto_list = []
  for noun in nouns:
      noun = nlp(noun)
      for item in items:
          if(item.similarity(noun) > 0.7):
              auto_list.append(noun.text)
  return list(set(auto_list))
```

```python
decor_auto = generate_auto_list(decor_nlp)
accessories_auto = generate_auto_list(accessories_nlp)
kitchenware_auto = generate_auto_list(kitchenware_nlp)
furniture_auto = generate_auto_list(furniture_nlp)
stationary_auto = generate_auto_list(stationary_nlp)
christmas_auto = generate_auto_list(christmas_nlp)
toys_games_auto = generate_auto_list(toys_games_nlp)
```

```python
def contains(list, row):
  for word in list:
    if word in row['description'] :
      return True
  return False
```

```python
def assign_tags(row):
  tags = ''
  if contains(decor_auto,row):
    tags += ' decor'
  if contains(accessories_auto,row):
    tags += ' accessories'
  if contains(kitchenware_auto,row):
    tags += ' kitchenware'
  if contains(furniture_auto,row):
    tags += ' furniture'
  if contains(stationary_auto,row):
    tags += ' stationary'
  if contains(christmas_auto,row):
    tags += ' christmas'
  if contains(toys_games_auto,row):
    tags += ' toys_games'
  if len(tags) > 1:
    return tags
  else:
    return None

unique_desc_df['tags'] = unique_desc_df.apply(assign_tags, axis=1)
unique_desc_df.sample(5)
```

```python
#check the equality of the variances with Levine test
# Null hypothesis - variances are the same, Alt hypothesis - variances are different
#Alpha = 0,05
statistic, pvalue = st.levene(only_order_customers['total_price'], several_order_customers['total_price'])
statistic, pvalue
```

The variances are different.

```python
#check the normality of the samples
#Null hypothesis - distribotion is normal, Alt hypothesis - distribotion is not normal
#Alpha = 0,05

result_one_order=st.shapiro(only_order_customers['total_price'])
result_several_orders=st.shapiro(several_order_customers['total_price'])
print(result_one_order,result_several_orders)
```

Since the p-values are lees than alpha we reject null hypothesis - distributions of the samples are not normal. So we have to run Mann Whitney U test.

```python
#Null hypothesis - The average revenue from the clients who purchased once doesn't differs 
#from the average revenue from the clients who made multiple orders.

alpha = 0.05 

statistic, pvalue = st.mannwhitneyu(only_order_customers['total_price'], several_order_customers['total_price'], alternative = 'less')

print('p-value:', pvalue)

if pvalue < alpha:
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```

This means that there is difference bitween average revenue from different groups of clients.
